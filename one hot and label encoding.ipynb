```python
import pandas as pd
data = {
    "town": [
        "monroe township", "monroe township", "monroe township", "monroe township", "monroe township",
        "west windsor", "west windsor", "west windsor", "west windsor",
        "robinsville", "robinsville", "robinsville", "robinsville"
    ],
    "area": [2600, 3000, 3200, 3600, 4000, 2600, 2800, 3300, 3600, 2600, 2900, 3100, 3600],
    "price": [550000, 565000, 610000, 680000, 725000, 585000, 615000, 650000, 710000, 575000, 600000, 620000, 695000]
}

df=pd.DataFrame(data)
print(df)
```

```python
df_encoded = pd.get_dummies(df.town).astype(int)
df_encoded
```

```python
df_combined=pd.concat([df,df_encoded],axis="columns")#concating /joining
df_combined
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dfle=df
dfle.town=le.fit_transform(dfle.town)
dfle
```
